#### Multiple Line - vergelijk ZO, NW en NO

```python

for naam, df in resultaten.items():
    if len(df['spatial_name'].unique()) == 4:
    #if df['value'].notna().any() and '_p' in naam:

        # Controleer of '_p' of '_r' in de indicatornaam staat en pas y-as daarop aan
        if '_p' in naam:
            y_min = 0
            y_max = 100 if df['value'].max() > 20 else 30
            is_percentage = True
        elif '_r' in naam:
            y_min = 1
            y_max = 10
            is_percentage = False
        else:
            # Gebruik standaardwaarden als geen '_p' of '_r' aanwezig is
            y_min = None
            y_max = None
            is_percentage = False 

        print(naam)
        multiple_line(
            df=df,
            x_col='temporal_date',
            y_col='value',
            main_area='Amsterdam', 
            y_min=y_min,
            y_max=y_max,
            color_func='#004699',
            is_percentage=is_percentage,
            output_path=f'../../05 reports/Alle stadsdelen/Zuidoost/{naam}.png' 
        )
```
